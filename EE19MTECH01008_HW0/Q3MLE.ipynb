```python
'''
Q3) Write program to perform MLE for 
(a) Binomial,
(b) Poisson,
(c) Exponential,
(d) Gaussian,
(e) Laplacian.

 and match with the true distribution
 
Author: Subhra Shankha Bhattacherjee
Roll: EE19MTECH01008
'''

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Number of initialized data points
N_points = 100000
#Number of columns in histogram
n_bins = 100
```

```python
#(a) Binomial

print("enter p for Binomial distribution (0<p<1)")
p = float(input())

true_dist = np.random.binomial(1,p,N_points)  #true distribution

MLE = np.mean(true_dist) #ML estimator

gen_bin_1 = np.random.binomial(1,MLE,N_points)   #generated distribution

fig, axs = plt.subplots(1, 2, sharey=True, tight_layout=True)
axs[0].hist(true_dist, bins=n_bins,alpha=0.5,color='r',label='true distribution')
axs[0].legend()

axs[1].hist(gen_bin_1, bins=n_bins,label="generated")
axs[1].legend()

fig.set_size_inches(10, 5, forward=True)
```

```python
#(b) Poisson

print("lambda for poisson(>=0)")
Lambda_1 = float(input())

true_dist = np.random.poisson(Lambda_1,N_points) #true distribution

Lambda = np.mean(true_dist) #ML estimator

gen_dist = np.random.poisson(Lambda,N_points)  #generated distribution

fig, axs = plt.subplots(1, 2, sharey=True, tight_layout=True)
axs[1].hist(gen_dist, bins=n_bins,label='generated')
axs[1].legend()

axs[0].hist(true_dist, bins=n_bins,alpha=0.5,color='r',label='true distribution')
axs[0].legend()

fig.set_size_inches(10, 5, forward=True) 
```

```python
#(c) Exponential

print("beta (1/lambda) for Exponential distribution(>=0)")
beta = float(input())

true_dist = np.random.exponential(beta,N_points) #true distribution

beta1 = np.mean(true_dist) #ML estimator

gen_dist = np.random.exponential(beta1,N_points)  #generated distribution

fig, axs = plt.subplots(1, 2, sharey=True, tight_layout=True)
axs[0].hist(true_dist, bins=n_bins,label='true distribution',color='r',alpha=0.5)
axs[0].legend()

axs[1].hist(gen_dist, bins=n_bins,label='generated')
axs[1].legend()

fig.set_size_inches(10, 5, forward=True)
```

```python
#(d) Gaussian

print("Enter the mean for ground truth gaussian")
mean = float(input())
print("Enter std dev for the ground truth gaussian(>0)")
std = float(input())

true_dist = np.random.normal(mean,std,N_points) #true distribution

mean = np.mean(true_dist) #ML Estimator
std_dev = np.std(true_dist)  #ML Estimator

gen_dist = np.random.normal(mean,std_dev,N_points)  #generated distribution

fig, axs = plt.subplots(1, 2, sharey=True, tight_layout=True)
axs[0].hist(true_dist, bins=n_bins,color='r',alpha=0.5,label="true distribution")
axs[0].legend()

axs[1].hist(gen_dist, bins=n_bins,label="generated")
axs[1].legend()

fig.set_size_inches(10, 5, forward=True)
```

```python
#(e) Laplacian.

print("Give Mu for Laplace")
Mu1 = float(input())
print("Give Beta for Laplace(>0)")
Beta1 = float(input())

true_dist = np.random.laplace(Mu1,Beta1,N_points) #ture distribution

mu = np.median(true_dist)  #ML estimator
beta = np.mean(np.abs(true_dist-mu))  #ML estimator

gen_dist = np.random.laplace(mu,beta,N_points) #generated distribution

fig, axs = plt.subplots(1, 2, sharey=True, tight_layout=True)
axs[0].hist(true_dist, bins=n_bins,label="true distribution",alpha=0.5)
axs[0].legend()

axs[1].hist(gen_dist, bins=n_bins, label="generated")
axs[1].legend()

fig.set_size_inches(10, 5, forward=True)
```
