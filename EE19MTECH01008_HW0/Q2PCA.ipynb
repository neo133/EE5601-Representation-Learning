```python
'''
Q2) Implement principal component analysis (PCA) to decorrelate data.
Author: Subhra Shankha Bhattacherjee
Roll: EE19MTECH01008
'''
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def normalize(x):
    x = x.astype('float')
    for i in range(len(x)):
        x[i] = x[i] - np.mean(x[i])
    print(np.mean(x[0]))
    return x
```

```python
pic = Image.open('scene.jpg')
img = (np.array(pic))
w,h,c = img.shape
print(w,h,c)
```

```python
a = img.reshape((w*h,c))
print(a.dtype)
a = a.astype('float')
```

```python
a = a.T
a.shape
```

```python
a = normalize(a)
```

```python
Cxx = (np.matmul(a,a.T)/(len(a[0])))
print(Cxx)
```

```python
evalues,evecs = np.linalg.eig(Cxx)
print(evalues,evecs)
```

```python
p = np.matmul(evecs.T,a)
p = p.T
p = p.reshape(img.shape)
p
```

```python
p = p.astype(int)
```

```python
plt.imshow(p)
p
```
